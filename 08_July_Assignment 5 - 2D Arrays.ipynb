```python
Question 1

Convert 1D Array Into 2D Array

You are given a 0-indexed 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with  m rows and n columns using all the elements from original.

The elements from indices 0 to n - 1 (inclusive) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (inclusive) should form the second row of the constructed 2D array, and so on.

Return an m x n 2D array constructed according to the above procedure, or an empty 2D array if it is impossible.

Example 1:
    
Input: original = [1,2,3,4], m = 2, n = 2

Output: [[1,2],[3,4]]

Explanation:The constructed 2D array should contain 2 rows and 2 columns.

The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.

The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

Solution:-  class Solution:
    def construct2DArray(self, original: List[int], m: int, n: int) -> List[List[int]]:
        # make sure we can create the matrix first (m * n == len(array))
        # iterate through m first creatig a new list to append to matrix
        # add the elments to the row (new list) by adding the current element in original
        # after iteration, return the matrix array
        # Time: O(M + N) Space: O(N)
        
        if m * n != len(original):
            return []
        
        matrix = [[] for i in range(m)]
        curr = 0
        
        for i in range(m):
            for j in range(n):
                matrix[i].append(original[curr])
                curr += 1
                
        return matrix

Time complexity: O(N + M) Space complexity: O(N)
```

```python
Question 2

You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase may be incomplete.

Given the integer n, return the number of complete rows of the staircase you will build.

Example 1:

Input: n = 5

Output: 2

Explanation: Because the 3rd row is incomplete, we return 2.

Solution:-  Approach
we first initialize two pointers left and right to 1 and n, respectively.
We then enter a while loop that continues as long as left is less than or equal to right.
In each iteration, we compute the kpoint k between left and right using integer division (//) to avoid floating-point precision errors.
We then calculate the sum curr of the first k natural numbers using the formula curr = k * (k + 1) // 2.
If curr is equal to n, we have found the maximum value of k and we can return k as the answer.
If curr is less than n, we update left to k + 1, since we know that all values less than k will also result in a sum less than n.
If curr is greater than n, we update right to k - 1, since we know that all values greater than k will result in a sum greater than n.
We continue until left is greater than right. Finally, we return right as the maximum value of k that satisfies the condition.

Complexity
Time complexity : O(logn)O(logn)O(logn)
Space complexity : O(1)O(1)O(1)

Code
class Solution:
    def arrangeCoins(self, n: int) -> int:
        l, r = 1, n
        while l <=r:
            k = (l+r)//2
            curr = k * (k + 1)//2
            if curr == n:
                return k
            elif curr < n:
                l = k + 1
            else:
                r = k - 1
        return r
```

```python
Question 3

Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.

Example 1:

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

Explanation: After squaring, the array becomes [16,1,0,9,100].

After sorting, it becomes [0,1,9,16,100].

Solution:-  Intuition
Given an integer array sorted in non-decreasing order, we need to return an array of the squares of each number sorted in non-decreasing order. Since the given array is sorted, the squares of negativenegativenegative numbers will be in decreasingdecreasingdecreasing order and the squares of positivepositivepositive numbers will be in increasingincreasingincreasing order. We can merge these two arrays of squares using a two−pointertwo-pointertwo−pointer approach.

Approach
We will start with two pointers, leftleftleft and rightrightright, both at the two ends of the given array. We will also create an empty result array called ′res′'res' 
′
 res 
′
 .
We will loop until the leftleftleft pointer is less than or equal to the rightrightright pointer.
If the element pointed by the leftleftleft pointer is negative, we will compare its absoluteabsoluteabsolute value with the element pointed by the rightrightright pointer.
If the absolute value of the leftleftleft element is greater than or equal to the rightrightright element, we will add the square of the leftleftleft element to the ′res′'res' 
′
 res 
′
  array and move the leftleftleft pointer one position to the right.
Otherwise, we will add the square of the right element to the ′res′'res' 
′
 res 
′
  array and move the rightrightright pointer one position to the left.
If the element pointed by the leftleftleft pointer is non-negative, we will breakbreakbreak out of the loop since all remaining elements will also be non-negative and their squares will be in increasingincreasingincreasing order.
We will loop through the remaining elements in the array from the rightrightright pointer to the leftleftleft pointer and append their squares to the ′res′'res' 
′
 res 
′
  array.
Since we appended the squares of the remaining elements from right to left, we need to reverse the ′res′'res' 
′
 res 
′
  array to get the squares in non−decreasingnon-decreasingnon−decreasing order.
We will return the ′res′'res' 
′
 res 
′
  array.

Complexity
Time complexity:
The given array is iterated twice - once using the two-pointer
approach and once to loop through the remaining elements. 
Therefore, the time complexity of this algorithm is O(n), where 'n'
is the length of the given array.

Space complexity:
We are creating an empty array to store the result. Therefore, 
the space complexity of this algorithm is O(n), where 'n' is 
the length of the given array.

Code

class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        res = []  # to store the result
        left, right = 0, len(nums)-1  # initialize left and right pointers to the two ends of the array
        
        # loop until left pointer is less than or equal to right pointer
        while left <= right:
            if nums[left] < 0:  
                # if the element pointed by left pointer is negative
                # compare absolute value of left element with the right element
                if abs(nums[left]) >= nums[right]:
                    res.append(nums[left]**2)  
                    # if absolute value of left element is greater than or equal
                    # to right element, add square of left element to the result array
                    left += 1  # move left pointer one position to the right
                else:
                    res.append(nums[right]**2)  
                    # if absolute value of left element is less than right element, add square of right
                    # element to the result array
                    right -= 1  # move right pointer one position to the left
            else:  
                # if the element pointed by left pointer is non-negative, break out of the loop
                break
        
        # loop through the remaining elements from right pointer to left pointer and 
        # append their squares to the result array
        for i in range(right, left-1, -1):
            res.append(nums[i]**2)
        
        res.reverse()  # reverse the result array to get the squares in non-decreasing order
        return res
```

```python
Question 4

Given two 0-indexed integer arrays nums1 and nums2, return a list answer of size 2 where:

- answer[0] is a list of all distinct integers in nums1 which are not present in nums2.
- answer[1] is a list of all distinct integers in nums2 which are not present in nums1.

Note that the integers in the lists may be returned in **any** order.

Example 1:

Input: nums1 = [1,2,3], nums2 = [2,4,6]

Output: [[1,3],[4,6]]

Explanation:

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

Solution:-  Intuition
Use HashSet to remove the duplicates and subtract the arrays.

Approach
convert the first array to set to remove the duplicates
convert the second array to set to remove the duplicates
subtract set1 from set2 to get the integers in the first array that are not present in the second array and vice versa. Return the result as a list
Complexity
Time complexity: O(n)
Space complexity: O(n)
Code
class Solution:
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        set1 = set(nums1)
        set2 = set(nums2)
        return [list(set1 - set2), list(set2 - set1)]



```

```python
Question 5

Given two integer arrays arr1 and arr2, and the integer d, return the distance value between the two arrays.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.

Example 1:

Input: arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

Output: 2

Explanation:

For arr1[0]=4 we have:

|4-10|=6 > d=2

|4-9|=5 > d=2

|4-1|=3 > d=2

|4-8|=4 > d=2

For arr1[1]=5 we have:

|5-10|=5 > d=2

|5-9|=4 > d=2

|5-1|=4 > d=2

|5-8|=3 > d=2

For arr1[2]=8 we have:

|8-10|=2 <= d=2

|8-9|=1 <= d=2

|8-1|=7 > d=2

|8-8|=0 <= d=2

Solution:- Intuition
The first hint that the problem is not immediately suitable is that numbers are allowed to be duplicates in both cases. This means we have an instance of duplication within a set range, which can be reduced, thus saving on time and memory space. To do this, we first turn array 2 in to the sorted list of the set of values in array 2. This means array 2 now only takes up size O(unique values in array 2) and is done in time O( n log n + N), where n log n is the sort process of the unique values in array 2 and N is the size of array 2 originally. We then also get the frequency of values in array 1 in time O(M) where M is the size of array 1. We then get the value of n as the length of the set form of array 2.

Now we have the arrays as unique forms with array 1 knowing the frequency and array 2 as a set. This was done in O(N + M + n log n) time.

From here, we can loop over the sorted keys of our frequency, using binary search to find our index of insertion for the key in n log n, and update as needed 1 time for a subset of M values, thus taking only O(m + n log n) for the value loop process.

Approach
Set array 2 as a sorted list of set of array 2
Set frequency of values as the Counter of array 1
Set n to the length of array 2 after processing
Set answer to 0
For value in the frequency keys sorted

get index for the value using bisect left passing array 2 and value
closest value is set to 1001 to start
if you are at an index greater than 0, check one behind
if you are at an index that is not the last index, check one where you are at
if closest after the above (they update closest) is now greater than d
increment answer by frequency of values at value
return answer when done

Complexity
Time complexity: O( (M + N) log N)
O( N + n log n) to recast array 2
at worst this O(N log N)
O(M) to get counters of array 1
O(min(M, m)) to loop over frequencies
in which we do O(log n) at best or O(log N) at worst work
total is O(M + N + n log n + m log n)
reduces to O(M + N + (m+n)log (n))
at best this is O(M + N)
at worst this is O((M+N) log N)

Space complexity: O(M)
We reuse the space in array 2 so no additional space there
We may even shrink it which is kind of nice
We use O(M) space in frequency of values
We use O(log M) space when sorting at worst
Total is then O(M)

Code

class Solution:
    def findTheDistanceValue(self, arr1: List[int], arr2: List[int], d: int) -> int:
        # sort the second array as a sorted list of unique array2 values 
        arr2 = sorted(list(set(arr2)))
        # frequency find the first array 
        frequency_of_values = collections.Counter(arr1)
        # get size of array 2 
        n = len(arr2)
        # set up answer storage 
        answer = 0
        # loop over each value in frequency keys sorted 
        for value in sorted(frequency_of_values.keys()) : 
            # get the index for this value in the second array 
            indice_for_value = bisect.bisect_left(arr2, value)
            # set closest out of range 
            closest = 1001
            # if indice is greater than 0, set closest to min of indice prior and value difference 
            if indice_for_value > 0 : 
                closest = min(closest, abs(arr2[indice_for_value-1] - value))

            # if indice is less than size of array 2 
            if indice_for_value < n : 
                # set closest to be min of self and arr2 at indice minus value 
                closest = min(closest, abs(arr2[indice_for_value] - value))

            # if closest is greater than d 
            if closest > d : 
                # increment by the frequency of the value in array 1 
                answer += frequency_of_values[value]
        # return answer when done 
        return answer 
```

```python
Question 6

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears once or twice, return an array of all the integers that appears twice.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

Example 1:

Input: nums = [4,3,2,7,8,2,3,1]

Output:

[2,3]

Solution:- # Idea: We notice that the problems states integers that consist in nums
# will be in the range [1, n]. We can exploit this factor to guarrentee a
# linear runtime and constant space solution.
class Solution:
    def findDuplicates(self, nums: List[int]) -> List[int]:
        res = []
        for num in nums:
            n = abs(num)
            if nums[n - 1] < 0:
                res.append(n)
            nums[n - 1] = -nums[n - 1]
        return res
The approach here is involves us making use of the factor that the numbers inside the list will be in the range [1, n].
Since that is the case, in a single pass, we can mark every (number - 1)'s corresponding index to be negative. If we respotted that marked index, this means the number is a duplicate and we append it to the returning list. This guarantees us a O(n) runtime and O(1) space.

Time complexity=O(n) , Space complexity=O(1)  
```

```python
Question 7

Suppose an array of length n sorted in ascending order is **rotated** between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

- [4,5,6,7,0,1,2] if it was rotated 4 times.
- [0,1,2,4,5,6,7] if it was rotated 7 times.

Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of unique elements, return *the minimum element of this array*.

You must write an algorithm that runs in O(log n) time.

Example 1:

Input: nums = [3,4,5,1,2]

Output: 1

Explanation:

The original array was [1,2,3,4,5] rotated 3 times.

Solution:- Intuition
Looking at the question, it's a sorted array, which is we have to use binary search.

Approach
Looking at some examples we find a pattern of what's happening. 
We could find that the solution is less than both the start and end position, as well as, least when compared with neighbouring element.
So we implement a binary search where the start and end values are changed picked on which has the minimum value. 
The minimum value side is got by checking for both the extremes from the mid values.

Complexity
Time complexity: O(logn)O(logn)O(logn)
Space complexity: O(1)O(1)O(1)

Code
class Solution:
    def findMin(self, nums: List[int]) -> int:
        s = 0
        e = len(nums) - 1

        while s <= e:
            mid = (s + e) // 2
            if nums[mid] > nums[s] and nums[s] < nums[e]:
                e = mid - 1
            elif nums[mid] > nums[e] and nums[e] < nums[s]:
                s = mid + 1
            elif mid + 1 <= e and nums[mid] > nums[mid + 1]:
                s = mid + 1
            elif mid - 1 >= s and nums[mid] > nums[mid - 1]:
                e = mid - 1
            else:
                return nums[mid]


```

```python
Question 8

An integer array original is transformed into a doubled array changed by appending twice the value of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original if changed is a doubled array. If changed is not a doubled array, return an empty array. The elements in* original *may be returned in **any** order*.

Example 1:

Input: changed = [1,3,4,2,6,8]

Output: [1,3,4]

Explanation: One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

Solution:- # Time Complexity: O(NlogN)
# Space Complextiy O(N)
# where N is the number of elements in `changed` 
class Solution:
    def findOriginalArray(self, changed):
        # use Counter to count the frequency of each element in `changed`
        cnt, ans = Counter(changed), []
        # if the length of the input is odd, then return []
        # because doubled array must have even length
        if len(changed) % 2: return []
        # sort in ascending order
        for x in sorted(cnt.keys()):
            # if the number of cnt[x] is greater than than cnt[x * 2]
            # then there would be some cnt[x] left
            # therefore, return [] here as changed is not a doubled array
            if cnt[x] > cnt[x * 2]: return []
            # handle cases like [0,0,0,0]
            if x == 0:
                # similarly, odd length -> return []
                if cnt[x] % 2:
                    return []
                else: 
                    # add `0` `cnt[x] // 2` times 
                    ans += [0] * (cnt[x] // 2)
            else:
                # otherwise, we put the element `x` `cnt[x]` times to ans
                ans += [x] * cnt[x]
            cnt[2 * x] -= cnt[x]
        return ans
```
